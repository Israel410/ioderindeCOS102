```python
## Python code for solving cubic equation
import numpy as np

print("There are Operations for solving quadratic, cubic and quartic equation")
choice=input("Select operation:1.Quadratic, 2.Cubic, 3.Quartic")
print("There are different operations in solving this equation")
if choice=="2":
    print("The cubic equation is ax^3+bx^2+cx+d=0")
    a=float(input("Enter the coeffecient of x^3, a: "))
    b=float(input("Enter the coefficient of x^2, b: "))
    c=float(input("Enter the coefficient of x, c: "))
    d=float(input("Enter the constant, d: "))

#Automatically calculate all 3 roots instead of guessing
    coeffs=[a,b,c,d]
    roots=np.roots(coeffs)
    print("\nThe roots are:")
    for i, r in enumerate(roots, 1):
        print(f"Root[i]: {r:.2f}")

if choice=="1":
    print("Quadratic equations is ax^2+bx+c")
    a=float(input("The coefficient of x^2, a:"))
    b=float(input("The coefficient of x, b:"))
    c=float(input("The constant c :"))
    determinant=b**2-(4*a*c)
    root1= (-b+(determinant**0.5))/(2*a)
    root2= (-b-(determinant**0.5))/(2*a)
    if determinant==0:
        print("The root is", root1)
    elif determinant>0:
        print(f"The roots are: {root1:.2f} and {root2:.2f}")
    if determinant<0:
        print("No true roots")


if choice=="3":
    print("The quartic equation with formula ax^4+bx^3+cx^2+dx+e")
    a=float(input("Enter the coeffecient of x^4, a: "))
    b=float(input("Enter the coefficient of x^3, b: "))
    c=float(input("Enter the coefficient of x^2, c: "))
    d=float(input("Enter the coefficient of x, d: "))
    e=float(input("Enter the constant value, e:"))

    
    coe=[a,b,c,d,e]
    roots=np.roots(coe)
    print("\nThe roots are:")
    for i, r in enumerate(roots, 1):
        print(f"Root[i]:{r:.2f}")
        
        
        
    

    
    
```
