```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

def verify_employee():
    # 1. Get user input from the boxes
    input_surname = surname_name_entry.get().strip()
    input_firstname = first_name_entry.get().strip()
    input_dept = department_entry.get().strip()

    try:
        # 2. Load the CSV file
        df = pd.read_csv('GIG-Logistics.csv')
        df.columns = df.columns.str.strip()

        # 3. Search for the employee (Case-insensitive)
        match = df[
            (df['SURNAME'].str.lower() == input_surname.lower()) & 
            (df['FIRST NAME'].str.lower() == input_firstname.lower()) & 
            (df['DEPARTMENT'].str.lower() == input_dept.lower())
        ]

        if not match.empty:
            # 4. Find coworkers in the same department
            is_same_dept = df['DEPARTMENT'].str.lower() == input_dept.lower()
            is_not_me = ~((df['SURNAME'].str.lower() == input_surname.lower()) & 
                          (df['FIRST NAME'].str.lower() == input_firstname.lower()))
            
            coworkers_df = df[is_same_dept & is_not_me]
            
            # Create the list of names
            names_list = (coworkers_df['FIRST NAME'] + " " + coworkers_df['SURNAME']).tolist()
            others_text = ", ".join(names_list) if names_list else "No other members"

            # 5. Show Success Message
            messagebox.showinfo("Access Granted", f"Welcome, {input_firstname}!\n\nDepartment Members:\n{others_text}")
        else:
            # 6. Show Failure Message
            messagebox.showerror("Access Denied", "Employee not found. Please check your details.")

    except FileNotFoundError:
        messagebox.showerror("File Error", "Could not find 'GIG-Logistics.csv'. Make sure it's in the same folder as this code.")

# --- GUI SETUP (This must be outside the function) ---

root = tk.Tk()
root.title("GIG Logistics - Verification")
root.geometry("400x450")

# Input for Surname
tk.Label(root, text="Surname:", font=("Arial", 10)).pack(pady=(20, 0))
surname_name_entry = tk.Entry(root, width=30)
surname_name_entry.pack(pady=5)

# Input for First Name
tk.Label(root, text="First Name:", font=("Arial", 10)).pack(pady=(10, 0))
first_name_entry = tk.Entry(root, width=30)
first_name_entry.pack(pady=5)

# Input for Department
tk.Label(root, text="Department:", font=("Arial", 10)).pack(pady=(10, 0))
department_entry = tk.Entry(root, width=30)
department_entry.pack(pady=5)

# The Verify Button
verify_btn = tk.Button(root, text="Verify Identity", command=verify_employee, bg="green", fg="white", font=("Arial", 10, "bold"))
verify_btn.pack(pady=30)

root.mainloop()
```
