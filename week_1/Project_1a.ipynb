```python
# Program to calculate Simple Interest
print("Calculate the Simple interest")

p = float(input("Input the Principal: "))     # The principal (P)
r = float(input("Input the rate in %: "))      # The rate (R) in %
t = float(input("Input the time in years: "))  # The time (T) in years

# Converting rate (R) in % into a fraction
rate = r / 100 

# The formula for Simple Interest (A)
a = p * (1 + (rate * t)) 

print(f"The simple interest is {a:.2f}")
```

## 1. Simple Interest
Simple interest is calculated only on the principal amount of a loan or deposit.

### The Formula
$$A = P \left(1 + \left(\frac{R}{100}\right)t\right)$$

### Flowchart Logic
* *Start*: Begin the program execution.
* *Input*: Receive values for Principal ($P$), Rate in % ($R$), and Time in years ($t$).
* *Process*: Divide the rate $R$ by 100 to convert it into a decimal fraction.
* *Calculation*: Apply the formula $A = P \times (1 + rate \times t)$.
* *Output*: Print the final Simple Interest amount $A$ formatted to two decimal places.
* *End*: Terminate the session.
