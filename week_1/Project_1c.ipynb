```python
# The python program for calculating Annuity plan
pmt=float(input("Enter the periodic payment is "))   # The periodic payment
r=float(input("Enter the annual rate in % "))   # The annual rate 
n=float(input("Enter the compounding frequency ")) # The compounding frequency
t=float(input("Enter the time in years ")) # The time in years
a=pmt*((1+(r/n)**(n*t))-1)/(r/n)  # The annuity plan formula
print(f"The annuity plan is {a:.2f}")

```

## 3. Annuity Plan
An annuity plan is a series of regular payments made over a specific period of time.

### The Formula
$$A = PMT \times \frac{\left[\left(1 + \frac{R}{n}\right)^{nt} - 1\right]}{\left(\frac{R}{n}\right)}$$



### Mathematical Logic for Python
To raise the formula to the power of $nt$ in Python, we use the ** operator. The logic is structured as follows:
1. Calculate the periodic rate: i = R / n.
2. Calculate the total periods: periods = n * t.
3. Apply the growth factor: (1 + i) ** periods.
4. Finalize the calculation by dividing the growth by the periodic rate and multiplying by the payment ($PMT$).
