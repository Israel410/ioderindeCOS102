```python
# Python Program for solving for Compound Interest
p=float(input("Enter the principal "))  # The principal
r=float(input("Enter the rate in % "))  # The rate in %
t=float(input("Enter the time in years ")) # The time in years
n=float(input("Enter the compounding frequency")) # The compounding frequency
a=p*((1+(r/n))**(n*t)) # Formula for compound interest
print(f"The Compound interest is {a:.2f}")


```

## 2. Compound Interest
Compound interest is the interest calculated on the initial principal, which also includes all of the accumulated interest from previous periods.

### The Formula
$$A = P \left(1 + \frac{R}{n}\right)^{nt}$$

### Variable Glossary
| Variable | Description |
| :--- | :--- |
| *$P$* | *Principal*: The initial amount of money. |
| *$R$* | *Annual Interest Rate*: The decimal representation of the rate. |
| *$n$* | *Compounding Frequency*: The number of times interest is applied per year. |
| *$t$* | *Time*: The number of years the money is invested. |
