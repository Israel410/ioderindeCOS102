```python
import pandas as pd

df=pd.read_csv('nyc bnb.csv')

df
```

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['name']=df['name'].fillna('No name')
```

```python
df.isnull().sum()
```

```python
df['host_name']=df['host_name'].fillna('No name')
```

```python
mode=df['last_review'].mode()

df['last_review']=df['last_review'].fillna(mode)
```

```python
df.isnull().sum()
```

```python
mean=df['reviews_per_month'].mean()

df['reviews_per_month']=df['reviews_per_month'].fillna(mean)
```

```python
df.isnull().sum()
```

```python
df['last_review']=pd.to_datetime(df['last_review'], errors='coerce')
```

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as pIt
import seaborn as sns


pIt.figure(figsize=(12,6))

# Order by count so the bars look organised

sns.countplot(data=df, x='neighbourhood_group',
order=df['neighbourhood_group'].value_counts().index, palette='viridis')

pIt.title('Number of Listings per Neighbourhood Group')
pIt.xlabel('Borough')
pIt.ylabel('Total Listings')
pIt.show()


```

Insights:
1. Manhattan and Brooklyn are dominating the chart
2. Areas with very few listings might be "emerging markets" or areas with stricter local rental laws


```python
pIt.figure(figsize=(10,8))

# We use latitude/longitude as coordinates and color by neighbourhood
sns.scatterplot(x='longitude', y='latitude', hue='neighbourhood_group', data=df, s=10, alpha=0.5)

pIt.title('Geographicall Distribution of Airbnb Listings')
pIt.xlabel('Longitude')
pIt.ylabel('Latitude')
pIt.legend(title='Borough', bbox_to_anchor=(1.05,1), loc='upper left')
pIt.show()
```

Insight
   1. Borders: You can actually see "shape" of the city and its islands just from the listing dots
   2. Concentration: You'll see where the clusters are thickest(usually downtown centers) versus where they thin out 

```python
# Filtering to listings under $500 so the "violin" shape isn't stretched out by outliers

sub_df=df[df['price']<500]

pIt.figure(figsize=(10, 6))
sns.violinplot(x='room_type', y='price', data=sub_df, palette='muted')


pIt.title('Price Distribution by Room Type')
pIt.xlabel('Room Type')
pIt.ylabel('Price($)')
pIt.show()
```

The insights:
    
    1. Price Probability: The "thickest" part of the violin shows where you are most likely to find a price
    
    2. Range: You'll notice "Entire homes" usually have a much taller (wider range) violin than "Shared rooms," which are usually very compressed at
    the bottom
